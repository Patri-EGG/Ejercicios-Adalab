Convierto los valores de la columna a minusculas para evitar errores:

```python
df_mer['product_color'] = df_mer['product_color'].str.lower()
```

```python
productos_verdes = df_mer[df_mer['product_color'] == 'green']
cantidad_productos_verdes = productos_verdes.shape[0] # Uso .shape[0]para contar el número de filas en el DataFrame productos_verdes

print(f'la cantidad de productos color verde  es:\n {cantidad_productos_verdes}')
```

Usamos .isin() porque son varios valores:

```python
productos_vr = df_mer[df_mer['product_color'].isin(['green', 'red'])]
cantidad_productos_vr = productos_vr.shape[0]
print(f'la cantidad de productos color verde y rojo es:\n {cantidad_productos_vr}')
```

Busco que columnas pueden tener el valor 'M':

```python
columnas_con_m = []
for columna in df_mer.columns:
    try:
        if df_mer[columna].astype(str).str.contains('M').any():
            columnas_con_m.append(columna)
    except AttributeError:
        # cuando la columna no puede ser convertida a string.
        continue

print(f"Columnas con 'M': {columnas_con_m}")
```

Uso 'product_variation_size_id':

```python
productos_m15 = df_mer[(df_mer['product_variation_size_id'] == 'M') & (df_mer['price'] > 15)]
productos_m15.sample(2)
```

```python
cantidad_m15 = productos_m15.shape[0]
print(f'la cantidad de productos talla "M" de precio mayor a 15 es:\n {cantidad_m15}')
```

```python
precio_1_2 = df_mer[(df_mer['price'] >= 1) & (df_mer['price'] <= 2)]
cantidad_precio_1_2 = precio_1_2.shape[0]
print(f' La cantidad de productos con precio entre 1-2 es: \n {cantidad_precio_1_2}')
```

Uso Regex para seleccionar titulos que contengan digitos:

```python
productos_con_digitos = df_mer[df_mer['title'].str.contains(r'\d')]
productos_con_digitos['title'].head()
```
