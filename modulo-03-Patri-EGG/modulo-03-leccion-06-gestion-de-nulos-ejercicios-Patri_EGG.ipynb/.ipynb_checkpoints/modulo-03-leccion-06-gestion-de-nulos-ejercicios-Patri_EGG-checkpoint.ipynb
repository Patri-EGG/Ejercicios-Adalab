```python
df = pd.read_csv('df-rating-insight-leccion-05.csv', index_col = 0)
df.head(1)
```

```python
df = df.drop(columns='Unnamed: 0', axis=1)
```

```python
# Identifico columnas con valores nulos y su respectivo conteo
null_counts = df.isnull().sum()
null_columns = null_counts[null_counts > 0]

# Proporción de valores nulos por columna
null_proportions = (null_columns / len(df)) * 100

# Creo un DataFrame para visualizar el conteo y la proporción de nulos por columna
null_stats = pd.DataFrame({
    'Column': null_columns.index, 
    'Null Count': null_columns.values, 
    'Null Proportion (%)': null_proportions.values
})
null_stats.sort_values('Null Proportion (%)', ascending=False)  # Ordenar por proporción de nulos de mayor a menor
```

```python
#  Eliminao columna con alta proporción de nulos
df.drop(columns='merchant_profile_picture', inplace=True)
```

```python
# Imputo valores nulos en las columnas de rating con 0
rating_columns = ['rating_five_count', 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']
df[rating_columns] = df[rating_columns].fillna(0)
```

```python
# Imputar valores nulos en 'product_color' usando la moda
imputer_mode = SimpleImputer(strategy='most_frequent')
df['product_color'] = imputer_mode.fit_transform(df[['product_color']]).ravel()

# Imputar valores nulos en 'product_variation_size_id' usando la moda
df['product_variation_size_id'] = imputer_mode.fit_transform(df[['product_variation_size_id']]).ravel()
```

```python
# Uso un valor constante para 'origin_country'
imputer_constant = SimpleImputer(strategy='constant', fill_value='Unknown')
df['origin_country'] = imputer_constant.fit_transform(df[['origin_country']]).ravel()
```

```python
# Imputo 'merchant_name' usando la categoría más frecuente
df['merchant_name'] = imputer_mode.fit_transform(df[['merchant_name']]).ravel()
```

```python
# Verifico la imputación y la ausencia de nulos
df.isnull().sum()
```

```python
df.to_csv('df-rating-insight-leccion-06.csv', index=False)
```
