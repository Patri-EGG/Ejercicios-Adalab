##  1.Utilizando la API de Balldontlie extrae la siguiente información.

El nombre, apellido y equipo de cada uno de los jugadores que tenemos en la NBA. Para eso tendrás que usar el siguiente endpoint. Extrae información de 200 jugadores. Devuelve los resultados en un DataFrame. Te deberá quedar algo parecido a esto:

```python
import pandas as pd
import requests
import re
```

```python
pd.set_option('display.max_columns', None)
```

Definir la clave de la API (es un identificador único que te permite acceder a la API)
Configurar los encabezados para la solicitud HTTP.
'Authorization' puede necesitar o no el prefijo 'Bearer ', dependiendo de los requisitos de la API.

```python
api_key = '166bbdf7-eae2-4379-9dc0-8dd9db1a7ded'
headers = {
    'Authorization': api_key                       
}
```

Definir la URL del punto final de la API al que deseas acceder.
Esta es la dirección específica proporcionada por la API donde se encuentra disponible la información solicitada.

 Definir la función para llamar a la API.

```python
def llamar_api(url, headers):
    # Realizar la solicitud GET pasando la URL y los encabezados.
    llamada = requests.get(url, headers=headers)
    # Imprimir el código de estado de la respuesta para verificar si fue exitosa.
    print(f"La llamada a la API nos ha dado una respuesta de tipo: {llamada.status_code}")
    # Si el código de estado es diferente de 200, imprimir la razón del fallo.
    if llamada.status_code != 200:
        print(f"El motivo por el que la llamada falló es {llamada.reason}")
    else:
        # Si la llamada fue exitosa, devolver la respuesta en formato JSON.
        return llamada.json()  

# Llamar a la función y almacenar la respuesta en una variable.    
url_jugadores = "http://api.balldontlie.io/v1/players"   
resultados_jugadores = llamar_api(url_jugadores, headers)
```

Aquí empieza el Ejercicio: Ejemplo con condicionales en caso que no estemos seguros de  como esta almacenada la información.

```python


# Verifica si recibiste datos antes de intentar crear el DataFrame.
if resultados_jugadores:
    # Procesa la respuesta JSON y crea una lista de diccionarios con la información de los jugadores.
    informacion_jugadores = [
        {'nombres': jugadores['first_name'], 'apellidos': jugadores['last_name'], 'equipo': jugadores['team']['full_name']}
        for jugadores in resultados_jugadores['data']
    ]
    # Crea un DataFrame de pandas con la lista de información de jugadores.
    df_equipos = pd.DataFrame(informacion_jugadores)
    #df_jugadores.index = df_jugadores.index + 1   #Si quiero que la lista empiece desde 1 en véz de 0
    print(df_equipos.head())  # Muestra las 5 primeras líneas del DataFrame para verificar su contenido.
else:
    print("No se pudo obtener la información de los jugadores.")   # Si no se recibieron datos, imprime un mensaje de error.
```
