```python
url_tablas = "https://es.wikipedia.org/wiki/Anexo:Medallero_de_los_Juegos_Ol%C3%ADmpicos_de_Tokio_2020"

res_tablas = requests.get(url_tablas)

print("La respuesta de la petición es:", res_tablas.status_code)
```

```python
sopa_tablas = BeautifulSoup(res_tablas.content, 'html.parser')
print(sopa_tablas.prettify())
```

```python
tablas = sopa_tablas.find_all("table")
print(f'Numero de tablas en la página: {len(tablas)}')
```

```python
print(tablas[0])
print(tablas[1])
print(tablas[2])
```

```python
mi_tabla = tablas[0]
print(mi_tabla)
```

```python
lista_encabezados = mi_tabla.find_all("th")
print(lista_encabezados)
```

```python
encabezados_medallero = [columna.text for columna in lista_encabezados]
print("Los encabezados de la tabla son:", encabezados_medallero)
```

```python
filas_medallero = mi_tabla.find_all("tr")
print("La cantidad de filas en mi_tabla son:", len(filas_medallero))
print(filas_medallero[0]) # La primera no me vale porque es la información de los encabezados.
print(filas_medallero[1]) 
```

```python
resultados_medallero = []   ## Creamos una lista para almacenar los resultados de las filas de mi_tabla.

for fila in filas_medallero[1:]:  #Iniciamos un bucle 'for' para iterar a través de la lista_medallero, pero comenzamos desde el segundo elemento de la lista.
    fila_texto = fila.text
    elementos_fila = fila_texto.split("\n")[1:-1]  # Para cada 'fila', extraemos el texto, lo dividimos en una lista usando '\n' como separador y eliminamos el primer y último elemento de la lista.
    resultados_medallero.append(elementos_fila)
    
print("Los resultados de iterar por la lista son:", resultados_medallero)
```

```python
 ## 'enumerate(resultados_medallero)' proporciona un contador (index_lista) y el valor (lista) mientras iteras sobre 'resultados_medallero'.
for index_lista, lista in enumerate(resultados_medallero):
    # 'enumerate(lista)' proporciona un contador (index_elemento) y el valor (elemento) mientras iteras sobre cada 'lista' dentro de 'resultados_medallero'.
    for index_elemento, elemento in enumerate(lista):
        try:
            # Se eliminan caracteres como '%' y se cambian puntos por comas para el formato numérico correcto.
            resultados_medallero[index_lista][index_elemento] = elemento.replace("\xa0", "").replace(".", "").replace(",", ".")
        except:
            # Si la conversión falla (por ejemplo, si 'elemento' contiene texto que no se puede convertir a flotante),
            # se mantiene el valor original del 'elemento' en la 'lista'.
            resultados_medallero[index_lista][index_elemento] = elemento
            
print("Los resultados después de limpiar los datos son: ", resultados_medallero)
 
```

### Aquí creamos una función con todo el código creado para unificar todo.

```python

def sacar_tabla_medallero(url):
    res_medallero = requests.get(url)
    sopa_medallero = BeautifulSoup(res_medallero.content, 'html.parser')
    tabla_medallero = sopa_medallero.find_all("table")[0]
    filas = tabla_medallero.find_all("tr")
    
    encabezados = [th.get_text(strip=True) for th in filas[0].find_all("th")]  # strip=True elimina caracteres y espacios indeseados provenientes del  HTML.
    datos = [[td.get_text(strip=True) for td in fila.find_all("td")] for fila in filas[1:] if fila.find_all("td")]
    
    return pd.DataFrame(datos, columns=encabezados)

# Uso de la función
df_medallero = sacar_tabla_medallero(url_tablas)

# Mostrar el DataFrame y hacer la exploración básica
print(df_medallero.head())
print("El número de filas y columnas es:", df_medallero.shape)
print("Tipos de datos en cada columna:\n", df_medallero.dtypes)
print("Valores únicos en la columna de País:\n", df_medallero["País"].unique())
print("Conteo de valores en la columna de País:\n", df_medallero["País"].value_counts())
print("Cantidad de valores no nulos en cada columna:\n", df_medallero.notnull().sum())
```
